# file_identification_script

```python
import os
import pandas as pd
import cv2
import PyPDF2 


f = []
t = []
x = []
y = []
z = []



input_dir = r'C:\Users\majidmm\Documents\AMXL\INPUT_FILES\all_input_files\\'


for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename)
            
            pdfFileObj = open(input_dir+filename, 'rb') 
  
            pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    
            pageObj = pdfReader.getPage(0) 
  
            pageObj.extractText()
    
#             print(pageObj.extractText())
  
            pdfFileObj.close() 
    
            t.append(pageObj.extractText())
        
            x.append('US DOT' in (pageObj.extractText()))
            y.append('District number' in (pageObj.extractText()))
            z.append('CUSTOMER REFERENCE NUMBER' in (pageObj.extractText()))
                    
            print('-='*20)
            
        except:
            continue
            
df = pd.DataFrame(list(zip(f, t, x, y, z)),columns=['file_Name','full_text','pattern_1','pattern_2','pattern_3'])
# df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\magma.csv",index=False)
pattern_1=df[df['pattern_1']== True]
pattern_1.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p1.csv",index=False)

pattern_2=df[df['pattern_2']== True]
pattern_2.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p2.csv",index=False)

pattern_3=df[df['pattern_3']== True]
pattern_3.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p3.csv",index=False)
```

# file_move_script

```python
import csv
import os
import shutil

def main():
    csv_file = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\magma_1.csv"
    from_folder = r"C:\Users\majidmm\Documents\AMXL\INPUT_FILES\all_input_files"
    to_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_p1,2,3\pattern_3"

    with open(csv_file, 'r', newline='') as f:
        reader = csv.reader(f)

        for row in reader:
            file_name = row[0]

            from_filename = os.path.join(from_folder, file_name)
            to_filename = os.path.join(to_folder, file_name)

            try:
                shutil.copy(from_filename, to_filename)
                print("Moved - '{}' -> '{}'".format(from_filename, to_filename))
            except shutil.Error as e:
                print("Failed - '{}' -> '{}'".format(from_filename, to_filename))

main()
```

# format_3_image_converter

```python
from PIL import Image

from pdf2image import convert_from_path
import os

from pdf2image.exceptions import (
 PDFInfoNotInstalledError,
 PDFPageCountError,
 PDFSyntaxError
)

input_dir = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_p1,2,3\pattern_3\\"

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:         
            PDF_PATH = input_dir+filename
            pages = convert_from_path(PDF_PATH, first_page=1, last_page=1,dpi=300,thread_count=1,fmt='jpg')
            for page in pages:
                page.save(filename.rstrip('.pdf') + ".jpg")
                print(filename)                            
        except:
            continue
```
