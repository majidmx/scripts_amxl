
# image_converter

```python
import os
import tempfile
from pdf2image import convert_from_path

from pdf2image.exceptions import PDFPageCountError



input_dir=r'C:\Users\majidmm\Documents\AMXL\INPUT_FILES\all_input_files\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        if filename.endswith(".pdf"):
            with tempfile.TemporaryDirectory() as path:
                pdf_path=input_dir+filename
                images_from_path = convert_from_path(pdf_path, output_folder=path, fmt= 'png',last_page=1, first_page =0,dpi=500,thread_count=1,poppler_path=r'C:\Users\majidmm\Downloads\poppler-0.68.0\bin\\')
                base_filename  =  os.path.splitext(os.path.basename(filename))[0] + '.png'     
                save_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\images_all\\'
                for page in images_from_path:
                    page.save(os.path.join(save_dir, base_filename), 'PNG')
                    print(filename)
                    continue
        else:
            continue
```

# file_identification_script

```python
import os
import pandas as pd
import cv2
from PIL import Image
import pytesseract
from pytesseract import Output


tesseract_exe_path_installation="C:\\Users\\majidmm\\AppData\\Local\\Tesseract-OCR\\tesseract.exe"
pytesseract.pytesseract.tesseract_cmd=tesseract_exe_path_installation

f = []
t = []
x = []
y = []
z = []
w = []



input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\images_all\\'


for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:          
            print(filename)
#             f_name = os.path.basename(input_dir).split('.')[0]
#             f.append(f_name.split()[0])
            f.append(filename)
            img = cv2.imread(input_dir+filename)
            custom_config = r' --psm 6'
            text = pytesseract.image_to_string(img,config=custom_config,output_type=Output.STRING)
            t.append(text.lower())
        
            x.append('us dot' in text.lower())
            y.append('district number' in text.lower())
            z.append('reference number' in text.lower())
            w.append('penske location' in text.lower())

                    
            print('-='*20)
            
        except:
            continue
            
df = pd.DataFrame(list(zip(f, t, x, y, z, w)),columns=['file_Name','full_text','pattern_1','pattern_2','pattern_3','pattern_4'])
pattern_1=df[df['pattern_1']== True]
# pattern_1['file_Name']=pattern_1['file_Name'].str.replace('png','pdf')
# pattern_1['file_Name']=pattern_1['file_Name']+'pdf'
pattern_1['file_Name'].str.replace('png','pdf').to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p1.csv",index=False, header=False)

pattern_2=df[df['pattern_2']== True]
pattern_2['file_Name']=pattern_2['file_Name']+'.pdf'
pattern_2['file_Name'].to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p2.csv",index=False, header=False)

pattern_3=df[df['pattern_3']== True]
pattern_3['file_Name']=pattern_3['file_Name']+'.png'
pattern_3['file_Name'].to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p3.csv",index=False, header=False)

pattern_4=df[df['pattern_4']== True]
pattern_4['file_Name']=pattern_4['file_Name']+'.pdf'
pattern_4['file_Name'].to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p4.csv",index=False, header=False)
```

# file_move_script

```python
import csv
import os
import shutil

def main():
    
    

    with open(csv_file, 'r', newline='') as f:
        reader = csv.reader(f)

        for row in reader:
            file_name = row[0]

            from_filename = os.path.join(from_folder, file_name)
            to_filename = os.path.join(to_folder, file_name)

            try:
                shutil.copy(from_filename, to_filename)
                print("Moved - '{}' -> '{}'".format(from_filename, to_filename))
            except shutil.Error as e:
                print("Failed - '{}' -> '{}'".format(from_filename, to_filename))

csv_file = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p1.csv"
from_folder = r"C:\Users\majidmm\Documents\AMXL\INPUT_FILES\all_input_files"
to_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_1"
main()
csv_file = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p2.csv"
from_folder = r"C:\Users\majidmm\Documents\AMXL\INPUT_FILES\all_input_files"
to_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_2"
main()
csv_file = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p3.csv"
from_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\images_all"
to_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_3"
main()
csv_file = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\file_type\p4.csv"
from_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\all_input_files"
to_folder = r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_4"
main()
```
