# format_1

```python
import os
import pandas as pd
import cv2
import camelot


f = []
v = []
v1 = []
m = []
x = []
y = []
z = []
s1=[]
s2=[]
s3=[]


input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_1\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
            print(filename)
            f.append(filename)
            tables = camelot.read_pdf(input_dir + filename,pages='1,2',flavor='lattice')
            a=tables[0].df
            b=tables[1].df
            
            vehicle_number = a.iloc[25, 5].strip()
#             print(vehicle_number)
            v.append(vehicle_number)
    
            vehicle_number_2 = a.iloc[25, 4].strip()
#             print(vehicle_number)
            v2=vehicle_number_2.strip('Vehicle Number')
            vx=v2.strip()
            v1.append(vx)
            
            mileage_rate = a.iloc[17,1].strip()
            m.append(mileage_rate.strip('$'))
            
            mileage_units=a.iloc[17, 2]
#             print(mileage_units)
            x.append(mileage_units)
        
            pick_date = a.iloc[14, 4].strip()
            pick_date.strip('Pick Date/Time')
#             print(pick_date.strip('Pick Date/Time'))
            y.append(pick_date.strip('Pick Date/Time'))
            
            return_date = a.iloc[15, 4].strip()
            return_date.strip('Return Date/Time')
#             print(return_date.strip('Return Date/Time'))
            z.append(return_date.strip('Return Date/Time'))
            
            try:
                sales_tax1=b.iloc[0,1].strip()
                s1.append(sales_tax1)
            except IndexError:
                s1.append('')
            
            try:
                sales_tax2=b.iloc[1,1].strip()
                s2.append(sales_tax2)
            except IndexError:
                s2.append('')
            
            try:
                sales_tax3=b.iloc[2,1].strip()
                s3.append(sales_tax3)
            except IndexError:
                s3.append('')
            
            
            print('-='*20)
            
df = pd.DataFrame(list(zip(f, v, v1, x, y, z, m, s1, s2, s3)),columns=['file_Name','VIN Number x','VIN Number y','Claimed Miles','Rental Start date','Rental End date','Mileage Rate','s1','s2','s3'])
df['VIN Number']=df['VIN Number x']+df['VIN Number y']
df['s1']=df['s1'].str.strip('$')
df['s2']=df['s2'].str.strip('$')
df['s3']=df['s3'].str.strip('$')


df['Sales Tax']=df['s1']+df['s2']+df['s3']
# df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p1_end.csv",index=False,columns=('file_Name','VIN Number','Mileage Rate','Claimed Miles','Rental Start date','Rental End date','s1','s2'))
df
```

```python
import os
import pandas as pd
import cv2
import camelot


f = []
v = []
x = []
y = []
z = []

input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_2\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename)
            tables = camelot.read_pdf(input_dir + filename,,pages='3',flavor='stream',split_text= True,edge_tol=500)
            a=tables[0].df
            
            vehicle_number = a.iloc[6, 2]
#             print(vehicle_number)
            v.append(vehicle_number)

            pick_date = a.iloc[10, 2]
#             print(pick_date.strip('Pick Date/Time'))
            x.append(pick_date.strip('Pick Date/Time'))

            return_date = a.iloc[11, 2]
#             print(return_date.strip('Return Date/Time'))
            y.append(return_date.strip('Return Date/Time'))

            mileage_units=a.iloc[20, 2]
#             print(mileage_units)
            z.append(mileage_units)

            period_billed = a.iloc[13, 2]
#             print(period_billed)
            w.append(period_billed)
            
        except:
            continue


df = pd.DataFrame(list(zip(f, v, x, y, z, w)),columns=['file_Name','VIN Number','Rental Start date','Rental End date','Claimed Miles','period_billed'])
df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p2_end.csv",index=False)
```

```python
from PIL import Image
import cv2 
import pytesseract
import re 
from pytesseract import Output
import pandas as pd
import os


tesseract_exe_path_installation="C:\\Users\\majidmm\\AppData\\Local\\Tesseract-OCR\\tesseract.exe"
pytesseract.pytesseract.tesseract_cmd=tesseract_exe_path_installation



f = []
v = []
m = []
x = []
y = []
z = []
s1 = []
s2 = []
s3 = []

input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_all_patterns\pattern_3\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename.rstrip('.png')+'.pdf')
            img = cv2.imread(input_dir+filename)
            custom_config = r' --psm 6'
            text = pytesseract.image_to_string(img,output_type=Output.STRING,config=custom_config)
            t=text.lower()
            
            left = 'hicle no:'
            right = 'customer information'
            vehicle_number = t[t.index(left)+len(left):t.index(right)]
            vehicle_number.strip(',')
            vehicle_number.strip(':')
#             print(vehicle_number)
            v.append(vehicle_number[0:7].strip())
            
            left = 'mileage rate'
            right = '/m ='
            mileage_rate = t[t.index(left)+len(left):t.index(right)]
            m.append(mileage_rate[-6:])
        
            
            left = 'out:'
            right = 'used:'
            pick_date = t[t.index(left)+len(left):t.index(right)]
#             print(pick_date[0:15])
            x.append(pick_date[0:10].strip())
            
            left = 'in:'
            right = 'st:'
            return_date = t[t.index(left)+len(left):t.index(right)]
#             print(return_date[0:15])
            y.append(return_date[0:10].strip())
            
            left = 'mileage rate'
            right = 'miles'
            mileage = t[t.index(left)+len(left):t.index(right)]
#             print(mileage)
            z.append(mileage.strip())
    
#             left = 'sales tax'
#             right = 'special tax'
#             sales_tax = t[t.index(left)+len(left):t.index(right)]
# #             print(sales_tax[-6:].strip())
#             s1.append(sales_tax[-6:].strip())
            
#             left = 'special tax =s'
#             right = '** special rate'
#             special_tax = t[t.index(left)+len(left):t.index(right)]
# #             print(special_tax.strip())
#             s2.append(special_tax.strip())

#             left = '1 week'
#             right = '** special rate'
#             week_tax = t[t.index(left)+len(left):t.index(right)]
# #             print(week_tax[-5:0].strip())
#             s3.append(week_tax[-5:0].strip())
            
        except:
            continue


df = pd.DataFrame(list(zip(f, v, z, x, y, m)),columns=['file_Name','VIN Number','Claimed Miles','Rental Start date','Rental End date','Mileage Rate'])
df['Claimed Miles']=df['Claimed Miles'].replace('o','0')
df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p3_end.csv",index=False,columns=('file_Name','VIN Number','Mileage Rate','Claimed Miles','Rental Start date','Rental End date'))
df
```
