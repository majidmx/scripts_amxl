```python
import os
import pandas as pd
import cv2
import camelot


f = []
v = []
v1 = []
x = []
y = []
z = []

input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_p1,2,3\pattern_1\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename)
            tables = camelot.read_pdf(input_dir + filename,flavor='lattice')
            a=tables[0].df
            
            vehicle_number = a.iloc[25, 5].strip()
#             print(vehicle_number)
            v.append(vehicle_number)
    
            vehicle_number_2 = a.iloc[25, 4].strip()
#             print(vehicle_number)
            v2=vehicle_number_2.strip('Vehicle Number')
            vx=v2.strip()
            v1.append(vx)
            
            mileage_units=a.iloc[17, 2]
#             print(mileage_units)
            x.append(mileage_units)
        
            pick_date = a.iloc[14, 4].strip()
            pick_date.strip('Pick Date/Time')
#             print(pick_date.strip('Pick Date/Time'))
            y.append(pick_date.strip('Pick Date/Time'))
            
            return_date = a.iloc[15, 4].strip()
            return_date.strip('Return Date/Time')
#             print(return_date.strip('Return Date/Time'))
            z.append(return_date.strip('Return Date/Time'))
            
            print('-='*20)
            
        except:
            continue


df = pd.DataFrame(list(zip(f, v, v1, x, y, z)),columns=['file_Name','VIN Number','VIN Number 2','Claimed Miles','Rental Start date','Rental End date'])
df
df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p1.csv",index=False)
```

# format_2

```python
import os
import pandas as pd
import cv2
import camelot


f = []
v = []
x = []
y = []
z = []

input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_p1,2,3\pattern_2\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename)
            tables = camelot.read_pdf(input_dir + filename,,pages='3',flavor='stream',split_text= True,edge_tol=500)
            a=tables[0].df
            
            vehicle_number = a.iloc[6, 2]
#             print(vehicle_number)
            v.append(vehicle_number)

            pick_date = a.iloc[10, 2]
#             print(pick_date.strip('Pick Date/Time'))
            x.append(pick_date.strip('Pick Date/Time'))

            return_date = a.iloc[11, 2]
#             print(return_date.strip('Return Date/Time'))
            y.append(return_date.strip('Return Date/Time'))

            mileage_units=a.iloc[20, 2]
#             print(mileage_units)
            z.append(mileage_units)

            period_billed = a.iloc[13, 2]
#             print(period_billed)
            w.append(period_billed)
            
        except:
            continue


df = pd.DataFrame(list(zip(f, v, x, y, z, w)),columns=['file_Name','VIN Number','Rental Start date','Rental End date','Claimed Miles','period_billed'])
df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p2.csv",index=False)
```

```python
from PIL import Image
import cv2 
import pytesseract
import re 
from pytesseract import Output

import os


tesseract_exe_path_installation="C:\\Users\\majidmm\\AppData\\Local\\Tesseract-OCR\\tesseract.exe"
pytesseract.pytesseract.tesseract_cmd=tesseract_exe_path_installation



f = []
v = []
x = []
y = []
z = []

input_dir = r'C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\files_dump_p1,2,3\pattern_3_1\\'

for root, dirs, filenames in os.walk(input_dir):
    for filename in filenames:
        try:
            print(filename)
            f.append(filename.strip('.jpg')+'.pdf')
            img = cv2.imread(input_dir+filename)
            custom_config = r'--psm 6'
            t = pytesseract.image_to_string(img,config=custom_config )
            
            left = 'VEHICLE NO:'
            right = 'CUSTOMER INFORMATION'
            vehicle_number = t[t.index(left)+len(left):t.index(right)]
#             print(vehicle_number)
            v.append(vehicle_number.strip())
            
            left = 'OUT:'
            right = 'USED:'
            pick_date = t[t.index(left)+len(left):t.index(right)]
#             print(pick_date[0:15])
            x.append(pick_date[0:15].strip())
            
            left = 'IN:'
            right = 'ST:'
            return_date = t[t.index(left)+len(left):t.index(right)]
#             print(return_date[0:15])
            y.append(return_date[0:15].strip())
            
            left = 'Mileage Rate'
            right = 'Miles'
            mileage = t[t.index(left)+len(left):t.index(right)]
#             print(mileage)
            z.append(mileage.strip())
            
        except:
            continue


df = pd.DataFrame(list(zip(f, v, x, y, z)),columns=['file_Name','VIN Number','Rental Start date','Rental End date','Claimed Miles'])
df.to_csv(r"C:\Users\majidmm\Documents\AMXL\OUTPUT_FILES\output_end\p3.csv",index=False)

```
